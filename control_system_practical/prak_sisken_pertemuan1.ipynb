```python
import numpy as np
import sympy as sp
import control as ct
```

NOMOR 1

```python
p_poly= np.array([1,-3, 2,-1])
q_poly= np.array([1, -1])

p_4= np.polyval((p_poly), 4)
q_4= np.polyval((q_poly), 4)
r= p_4*q_4
print("nilai r: ", r)
```

```python
#bentuk polinomial r(s)
r_poly = np.convolve(p_poly, q_poly)
print(r_poly)
```

```python
#menentukan akar polinomial r(s)
p_roots= np.roots(p_poly)
q_roots= np.roots(q_poly)
print(p_roots, "\n dan \n", q_roots)
```

```python
s= sp.symbols('s')
sysG_poly = (s**2+6*s+2)/((s+1)**3)
#print (sysG_poly)
#print("\n")
g= sp.apart(sysG_poly)
print("G(s) :", g)

sysH_poly = (s**2+4*s+5)/(s**3+6*s**2+11*s+6)
#print (sysH_poly)
#print("\n")
h= sp.apart(sysH_poly)
print("H(s) :", h)
```

```python
g_num = ([1, 6, 2 ])
g_den= ([1, 3, 3, 1])
sys1_tf= ct.TransferFunction(g_num, g_den)
print(sys1_tf)
z= ct.zeros(sys1_tf)
p= ct.poles(sys1_tf)
print(z)
print(p)
```

```python
h_num = ([1, 4, 5 ])
h_den= ([1, 6, 11, 6])
sys2_tf= ct.TransferFunction(h_num, h_den)
print(sys2_tf)
z1= ct.zeros(sys2_tf)
p1= ct.poles(sys2_tf)
print(z1)
print(p1)
```

```python
k=1
sys_zpk= ct.zpk(z, p, k)
print(sys_zpk)
```

```python
sys1_zpk= ct.zpk(z1, p1, k)
print(sys1_zpk)
```
