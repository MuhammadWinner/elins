```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

NOMOR 1

```python
#B
# Inisialisasi
K = [13]
R = 1000  # Ohm
C = 12e-7  # Farad
a = 1 / (R * C)
loop = 10
store_81 = {}

for i in range(loop):
    R = R + (R**2 / (R + (R * 10)))  
    C = C + 1e-2                     
    a = 1 / (R * C)
    print(f"R: {R:.2f}, C: {C:.6f}, a: {a:.2f}")
    
    Gs_81 = ct.TransferFunction(K[0], [1, a])
    store_81[i] = Gs_81

for i, tf in store_81.items():
    step_info81 = ct.step_info(tf)
    print (step_info81)
    t, y_step = ct.step_response(tf)
    plt.plot(t, y_step, label=f"sys {i+1}")

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Step Response for each R and C")
plt.grid(True)
plt.legend()
plt.show()

plt.figure()
for i, tf in store_81.items():
    ct.root_locus(tf, grid = True)

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Root Locust for each R and C")
plt.grid(True)
plt.legend()
plt.show()

```

```python
#C
# Inisialisasi
K = 13
R = 1000  # Ohm
C = 12e-7  # Farad
a = 1 / (R * C)
loop = 10
store_81 = {}

for i in range(loop):
    K =K+3
    print(f"K: {K:.2f}")
    
    Gs_81 = ct.TransferFunction([K], [1, a])
    store_81[i] = Gs_81

for i, tf in store_81.items():
    step_info81 = ct.step_info(tf)
    print (step_info81)
    t, y_step = ct.step_response(tf)
    plt.plot(t, y_step, label=f"sys {i+1}")

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Step Response for each K")
plt.grid(True)
plt.legend()
plt.show()

plt.figure()
for i, tf in store_81.items():
    ct.root_locus(tf, grid = True)

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Step Response for each R and C")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#B
K = 13
L = 2  # tetap
R = 1000  # Ohm
C = 12e-7  # Farad
loop = 15
store_81 = {}

for i in range(loop):
    
    wn = 1 / np.sqrt(L * C)
    xi = (R / (2 * L)) * np.sqrt(C / L)

    num = [K]
    den = [1, 2 * xi * wn, wn ** 2]
    Gs_81 = ct.TransferFunction(num, den)
    store_81[i] = Gs_81

    print(f"Iterasi {i+1}: K ={K:.2f}")

    K+=3
    
plt.figure()
for i, tf in store_81.items():
    t, y = ct.step_response(tf)
    plt.plot(t, y, label=f"Sys {i+1}")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Step Response (R & C berubah)")
plt.grid(True)
plt.legend()
plt.show()

plt.figure()
for i, tf in store_81.items():
    ct.root_locus(tf, grid=True)
plt.title(f"Root Locus - Sys {i+1}")
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.grid(True)
plt.show()
```

```python
#C
K = 13
L = 2  # tetap
R = 1000  # Ohm
C = 12e-7  # Farad
loop = 20
store_81 = {}

for i in range(loop):
    
    wn = 1 / np.sqrt(L * C)
    xi = (R / (2 * L)) * np.sqrt(C / L)

    num = [K]
    den = [1, 2 * xi * wn, wn ** 2]
    Gs_81 = ct.TransferFunction(num, den)
    store_81[i] = Gs_81

    print(f"Iterasi {i+1}: R = {R:.2f}, C = {C:.6e}, ξ = {xi:.4f}, ωn = {wn:.2f}")

    R += R**2 / (R + (R * 7))  
    C += 2e-8
    
plt.figure()
for i, tf in store_81.items():
    t, y = ct.step_response(tf)
    plt.plot(t, y, label=f"Sys {i+1}")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Step Response (R & C berubah)")
plt.grid(True)
plt.legend()
plt.show()

plt.figure()
for i, tf in store_81.items():
    ct.root_locus(tf, grid=True)
plt.title(f"Root Locus - Sys {i+1}")
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.grid(True)
plt.show()
```
