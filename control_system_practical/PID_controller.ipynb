```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

NOMOR 1

```python
#A
num_G = [1,20]
den_G = [1, 19, 80, -100]

Gs = ct.TransferFunction(num_G, den_G)
print(Gs)
ct.nyquist_plot(Gs)
plt.title("Nyquist plot")
plt.grid(True)
plt.show()
```

```python
#B
gm_Gs = 64.2 #dB
Fgm_Gs = 41.2 #rad/s
 
Kcr = ct.db2mag(gm_Gs)
print (f"gain margin (Kcr) = {Kcr:.2f}")
Pcr = 2*np.pi/Fgm_Gs
print((f"(Pcr) = {Pcr:.2f}"))
```

```python
#C
s = ct.TransferFunction.s
kp_Gs= 0.6 * Kcr
Ti_Gs= 0 * Pcr
Td_Gs= 0.125 * Pcr

Gc_Gs = kp_Gs*(1+ Td_Gs*s)
print(Gc_Gs)

GclC_Gs = ct.feedback (Gc_Gs*Gs, 1)
t, y= ct.step_response(GclC_Gs)
plt.figure()
plt.plot(t, y)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.grid()
plt.show()
```

```python
#D
s = ct.TransferFunction.s
kp_Gs= 0.45 * Kcr
Ti_Gs= Pcr/1.2
Td_Gs= 0 * Pcr

Gc_Gs = kp_Gs*(1+ 1/(Ti_Gs*s))
print(Gc_Gs)

GclD_Gs = ct.feedback (Gc_Gs*Gs, 1)
t, y= ct.step_response(GclD_Gs)
plt.figure()
plt.plot(t, y)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.grid()
plt.show()
```

```python
#E
s = ct.TransferFunction.s
kp_Gs= 0.6 * Kcr
Ti_Gs= 0.5 * Pcr
Td_Gs= 0.125 * Pcr
print(kp_Gs)
print(Ti_Gs)
print(Td_Gs)

Gc_Gs = kp_Gs*(1+ (1/(Ti_Gs*s)) + Td_Gs*s)
print(Gc_Gs)

GclE_Gs= ct.feedback (Gc_Gs*Gs, 1)
t, y= ct.step_response(GclE_Gs)
plt.figure()
plt.plot(t, y)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.grid()
plt.show()
```

```python
#f
C_info = ct.step_info(GclC_Gs)
print(C_info)

D_info = ct.step_info(GclD_Gs)
print(D_info)

E_info = ct.step_info(GclE_Gs)
print(E_info)
```
