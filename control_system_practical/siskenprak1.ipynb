```python
import numpy as np
#p(t) = t^3-t^2-4t+4
p_poly = np.array([1, -1, -4, 4])

#nilai polynomial pada t  = 4
p_4 = np.polyval (p_poly, 4)
print ("Nilai p(4) : ", p_4)

#akar polynomial
roots= np.roots(p_poly)
print ("Akar - akar polinomial :", roots)
```

```python
import sympy as sp

#mendefinisi variabel simbolik
t= sp.symbols ('t')

#representasi simbolis polinomial p(t) = t^3-t^2-4t+4
p_sym = t**3-t**2-4*t +4
print("Polynomial simbolis :", p_sym)
```

```python
#konversi polinomial dan simbolis 
p_poly_from_sym = sp.Poly(p_sym, t).all_coeffs()
print("Koefisien polinomial dari bentuk simbolis :", p_poly_from_sym)
```

```python
import control as ct
num = [1,2]
den= [1,4,3,0]

sys_tf = ct.TransferFunction(num, den)
print("Fungsi alih :", sys_tf)
```

model zpk

```python
#menentukan zero, pole
z= ct.zeros(sys_tf)
p = ct.poles(sys_tf)

print ("Zero:", z)
print ("pole:", p)

k = 1 # gain=1
sys_zpk=ct.zpk(z,p,k)#fungsi alih berdasar zero-pole-gain
print("Fungsi alih:", sys_zpk)
```

```python
#visualisasi lokasi zeri & pole pada s-plane
ct.pole_zero_plot(sys_zpk, grid=True)
```
